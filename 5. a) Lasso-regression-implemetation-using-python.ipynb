## Lasso Regression (L1- Regularization)


Lasso regression, short for Least Absolute Shrinkage and Selection Operator, is another type of linear regression that introduces regularization to the traditional least squares method. Like ridge regression, lasso regression aims to prevent overfitting and encourage simpler models by adding a penalty term to the cost function.

Cost function used is (1/2n)*(y-Xw)^2 + α |(slope)|

here we used (1/2n)*(y-Xw)^2 as last square error function

Xw is the predicted output

n is the number of smaples

α |(slope)| is the lasso penalyty term
