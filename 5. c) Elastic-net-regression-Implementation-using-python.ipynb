## Elastic net Regression

Elastic Net regression is a linear regression model that combines the penalties of both L1 (Lasso) and L2 (Ridge) regularization methods. It is used for variable selection and handling multicollinearity in regression analysis.

In linear regression, the goal is to find the coefficients that minimize the difference between the predicted values and the actual values of the dependent variable. Regularization techniques like Lasso (L1) and Ridge (L2) are introduced to prevent overfitting and improve the model's generalization to new data.

Elastic Net regression adds both L1 and L2 penalties to the linear regression cost function. The objective function for Elastic Net is a combination of the L1 and L2 regularization terms, controlled by two hyperparameters: alpha and l1_ratio.

L1 Regularization (Lasso): Encourages sparsity in the model by adding the absolute values of the coefficients to the cost function. Some coefficients may become exactly zero, effectively excluding the corresponding features from the model.

L2 Regularization (Ridge): Adds the squared values of the coefficients to the cost function, penalizing large coefficients. This helps to prevent multicollinearity by shrinking the coefficients towards zero.

Cost function is given by

1 / (2 * n_samples) * ||y - Xw||^2_2 + alpha * l1_ratio * ||w||_1 + 0.5 * alpha * (1 - l1_ratio) * ||w||^2_2

**1 / (2 * n_samples) * ||y - Xw||^2_2** --> Least sqaure error

**alpha * l1_ratio * ||w||_1** --> L1 Regularization
    
**0.5 * alpha * (1 - l1_ratio) * ||w||^2_2** --> L2 Regularization

If L1_Ratio is 0.5 then the elastic net regression cost function is the combination of both lasso and ridge penalty
